# INSCT自己提供的模拟数据集

```python
from ivis import Ivis
from sklearn.preprocessing import MinMaxScaler
from sklearn import datasets
import matplotlib.pyplot as plt
import hnswlib
import scanpy as sc
import pandas as pd
from tnn import *
##############################
# import tnn
# print(tnn.__version__)#version=0.0.2
##############################
adata=sc.read("/Users/xiaokangyu/Desktop/单细胞学习/单细胞数据集/simulation1/Simulation_data.h5")
# adata_umap = adata.copy()
# sc.pl.umap(adata,color=['batch','cells'])

X = adata.obsm["X_pca"]
model = Ivis(k = 50, model = 'szubert', verbose=0)
model.fit(X)
adata.obsm["X_ivis"] = model.transform(X)
sc.pl.embedding(adata, basis = 'ivis', color = ["batch", "cells"])

model = TNN(k=15, distance='pn', batch_size=64, n_epochs_without_progress = 10, verbose = 0)
model.fit(X = adata, batch_name='batch')
adata.obsm["X_tnn"] = model.transform(X = adata)
sc.pl.embedding(adata, basis = 'tnn', color = ["batch", "cells"])
```

# 4batch 4group run1

```python
import scanpy as sc 
from tnn import *

adata=sc.read("/Users/xiaokangyu/Desktop/单细胞学习/单细胞数据集/splatter_sim/4batch_4celltype.h5ad")
adata.obs["celltype"]=adata.obs["Group"].copy()
adata.obs["BATCH"]=adata.obs["Batch"].copy()

sc.pp.normalize_per_cell(adata,counts_per_cell_after=10000) ## same as INSCT
sc.pp.log1p(adata) # same as INSCT
sc.pp.highly_variable_genes(adata, n_top_genes=1000, subset = True) ## same as INSCT
sc.tl.pca(adata,svd_solver='arpack') # same as INSCT
#########################################
sc.pp.neighbors(adata)
sc.tl.umap(adata)
sc.pl.umap(adata,color=['Batch','Group'])
#########################################

model = TNN(k=20,verbose = 0) # set k=20，其余的保持默认
model.fit(X = adata, batch_name='Batch')
adata.obsm["X_tnn"] = model.transform(X = adata)
sc.pl.embedding(adata, basis = 'tnn', color = ["Group", "Batch"])


## 显示坐标刻度
fig, axs = plt.subplots(1, 2, figsize=(12,5),constrained_layout=True)
sc.pl.embedding(adata,basis="tnn",color="celltype", title="celltype", ax=axs[0], show=False)## 注意这个地方一定要show=False
sc.pl.embedding(adata,basis="tnn",color="BATCH", title="BATCH", ax=axs[1], show=False)## 注意这个地方一定要show=False,否则只能显示一张图片
axs[0].set_xscale('linear')#x轴设置刻度
axs[0].set_yscale('linear')#y轴设置刻度
axs[1].set_xscale('linear')#x轴设置刻度
axs[1].set_yscale('linear')#y轴设置刻度
plt.show()
```
